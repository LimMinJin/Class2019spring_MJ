```python
#TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

#Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

layer 하나 썼는데 node 10개

param#이 숫자 - 화살표의 갯수

y=ax+b  : a부분이 parameter

13개의 동그라미 + 1개의 동그라미 -> 14개

그 다음에 연결되는 hidden layer의 동그라미 수 10개 

14 * 10 = 140

그 다음, hidden layer에 동그라미 1을 하나 더 붙인다. (11개)

그 전 동그라미와는 연결될 필요가 없으나, 다음 단계의 동그라미와는 연결이 되어야 함

if 그 다음 단계 동그라미가 3개 --> 11 * 3 = 33개 

또 그 다음 동그라미가 1개면 

4 * 1 = 4

즉, 화살표를 만들 때에는 꼭 1을 붙여준다.

activation : 관계가 서로 non linear할 때 부여 -> 성능이 올라감

sigmoid, tanh, relu 부여 가능

regression일 때는 activation을 안 주고
> classification일 때는 
   
> sigmoid : binary classification

> softmax : multinary classification

사실 regression보다는 classification이 유용


```python
model.fit(x_train, y_train, epochs=1000)
```

epoch = 1 : data 전체를 한 번 돌리는 것

epoch = 1000 : data를 천 번 돌리는 것

돌릴수록 loss가 작아짐

```python
model.evaluate(x_test, y_test)
```
